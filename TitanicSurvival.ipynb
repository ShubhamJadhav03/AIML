```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
titanic = pd.read_csv("/content/drive/MyDrive/AIML/Decision_Tree/titanic (1).csv")
```

```python
titanic.head(5)
```

```python
titanic_survival = titanic.drop(columns=['PassengerId','Name', 'SibSp',	'Parch', 'Ticket', 	'Cabin', 'Embarked'])
```

```python
titanic_survival.head()
```

```python
target = titanic_survival.Survived
```

```python
X = titanic_survival.drop(['Survived'], axis="columns")
```

```python
X.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex = LabelEncoder()
```

```python
X['Sex'] = le_sex.fit_transform(X['Sex'])
X.dropna()
```

```python
from sklearn.model_selection import train_test_split
(x_train, x_test, y_train, y_test) = train_test_split(X, target, test_size= 0.2, random_state = 10)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
