```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/drive/MyDrive/AIML/Linear_Regression/homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area (sq ft)')
plt.ylabel('price ($)')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price) # Pass 'area' as a DataFrame
```

```python
reg.predict([[3400]]) # Pass the input as a 2D array
```

```python
area = pd.read_csv("/content/drive/MyDrive/AIML/Linear_Regression/areas.csv")
```

```python
price = reg.predict(area)
```

```python
area["price"] = price
```
