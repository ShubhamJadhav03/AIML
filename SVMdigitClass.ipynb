```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target'] = digits.target
df.head()
```

```python
digits.target_names
```

```python
df['target_digit'] = df.target.apply(lambda x : digits.target[x])
```

```python
df.target_digit
```

```python
X= df.drop(['target', 'target_digit'], axis = 'columns')
```

```python
y = df.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size = 0.2, random_state = 44)
```

```python
from sklearn.svm import SVC
model = SVC(kernel = 'rbf')
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
import numpy as np

# Assuming you have already trained your SVM model and named it 'model'

# Choose a sample from your test set (X_test)
sample_index = 24  # Choose any index from your test set
sample = X_test.iloc[sample_index].values.reshape(1, -1)  # Reshape to a 2D array

# Make the prediction
predicted_digit = model.predict(sample)[0]

# Print the predicted digit
print("Predicted digit:", predicted_digit)

# You can also print the actual digit for comparison
actual_digit = y_test.iloc[sample_index]
print("Actual digit:", actual_digit)
```
